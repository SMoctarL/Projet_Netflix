```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
print("Shape :", df.shape)
df.info()
```

```python
#  Vérifier combien il y a de valeurs manquantes par colonne (NA)
df.isna().sum()
```

```python
# Supprimer toutes les lignes qui contiennent AU MOINS un NA
df_clean = df.dropna()
```

```python
# Dimensions avant nettoyage
df.shape
```

```python
# Dimensions après nettoyage
df_clean.shape 
```

```python
#  Vérification des duplicats
df_clean.duplicated().sum()
```

```python

# Compter le nombre de films et séries
type_counts = df_clean['type'].value_counts()
print("Nombre de films et séries :")
print(type_counts)
# Explication : value_counts() compte combien de fois chaque type apparaît dans la colonne 'type'


# Calculer les proportions en pourcentage
type_proportions = df_clean['type'].value_counts(normalize=True) * 100
print("\nProportion de films et séries (%) :")
print(type_proportions.round(2))
# Explication : normalize=True calcule la proportion relative, *100 pour le pourcentage, round(2) pour arrondir à 2 décimales

#Commentaire simple
print("\nCommentaire :")
print(f"Netflix propose {type_counts['Movie']} films ({type_proportions['Movie']:.2f}%) et "
      f"{type_counts['TV Show']} séries ({type_proportions['TV Show']:.2f}%). "
      "On observe que la plateforme est légèrement orientée vers les films.")

```

```python
sns.countplot(data=df_clean, x="type")
plt.title("Répartition Films vs Séries")
# Afficher les valeurs sur chaque barre
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}',  # <-- converti en entier
                       (p.get_x() + p.get_width()/2., p.get_height()),
                       ha='center', va='bottom', fontsize=11)
plt.show()
```

```python
##Comparer nombre de films et séries par année

# Grouper par année et type de contenu, compter les occurrences
year_type_counts = df.groupby(['release_year', 'type']).size().unstack(fill_value=0)

# Afficher les premières lignes
print(year_type_counts)
```

```python
# Taille de la figure
plt.figure(figsize=(12,6))

# Tracer Films
plt.plot(year_type_counts.index, year_type_counts['Movie'], label='Films')

# Tracer Séries
plt.plot(year_type_counts.index, year_type_counts['TV Show'], label='Séries')

# Titre et labels
plt.title("Tendances Netflix par année : Films vs Séries", fontsize=14)
plt.xlabel("Année de sortie", fontsize=12)
plt.ylabel("Nombre de contenus", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Identifier les années où Netflix a ajouté beaucoup de contenu
df_clean["release_year"].value_counts().sort_index().plot(kind="bar", figsize=(15,5))
plt.title("Nombre de films/séries par année")
plt.xlabel("Année")
plt.ylabel("Nombre de titres")
plt.show()
```

```python
# Total contenus par année
total_per_year = year_type_counts.sum(axis=1)

# Années avec explosion : au-dessus de la moyenne + 1 écart-type
threshold_peak = total_per_year.mean() + total_per_year.std()
peak_years = total_per_year[total_per_year > threshold_peak].index.tolist()

# Années calmes : les 5 années avec le moins de contenus
calm_years = total_per_year.nsmallest(5).index.tolist()

# Type dominant par année
dominant_type = year_type_counts.apply(lambda row: 'Films' if row['Movie'] > row['TV Show'] 
                                       else ('Séries' if row['TV Show'] > row['Movie'] else 'Égal'), axis=1)

# Exemple de commentaire automatique
print("\nCommentaire :")
print(f"L’analyse montre que le nombre de contenus Netflix a progressivement augmenté après les années 2000, "
      f"avec des pics vers les années {peak_years[:3]}... "
      f"Les années calmes incluent {calm_years[:3]}... "
      f"Les films dominent légèrement le nombre total de contenus, mais on observe une augmentation régulière   des séries depuis 2010.")
```

```python
# les pays les plus représentés
df_clean["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 des pays les plus représentés")
plt.show()
```

```python
## Extraire tous les genres

#  Séparer les genres par virgule et "aplatir" la liste
all_genres = df_clean['listed_in'].str.split(', ').explode()

# Afficher les premières valeurs pour vérifier
print(all_genres.head())
```

```python
## Compter les genres les plus fréquents

# Compter la fréquence de chaque genre
genre_counts = all_genres.value_counts()

# Afficher les 10 genres les plus fréquents
print("Top 10 genres les plus fréquents :")
print(genre_counts.head(10))
# Catégorie la plus dominante
dominant_genre = genre_counts.idxmax()
dominant_count = genre_counts.max()

print(f"Le genre le plus fréquent est '{dominant_genre}' avec {dominant_count} occurrences.")

```

```python
# Top 10 genres
top10 = genre_counts.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top10.values, y=top10.index, color='skyblue')  # <-- utiliser color à la place de palette
plt.title("Top 10 des genres Netflix", fontsize=14)
plt.xlabel("Nombre de contenus", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.show()
```

```python
all_directors = df_clean['director'].str.split(', ').explode()
director_counts = all_directors.value_counts()
print("Top 10 réalisateurs les plus fréquents :")
print(director_counts.head(10))
```

```python
all_cast = df_clean['cast'].str.split(', ').explode()
cast_counts = all_cast.value_counts()
print("Top 10 acteurs/actrices les plus fréquents :")
print(cast_counts.head(10))
```

```python
num_unique_directors = all_directors.nunique()
num_unique_cast = all_cast.nunique()

print(f"\nNombre de réalisateurs uniques : {num_unique_directors}")
print(f"Nombre d'acteurs/actrices uniques : {num_unique_cast}")
print("\nCommentaire : Beaucoup de valeurs uniques et un casting très dispersé, ce qui est normal dans ce type de dataset.")
```

## 2.d simple analyse des Castings & Réalisateurs



```python
# --- TOP 10 RÉALISATEURS ---
director_counts = all_directors.value_counts().head(10).drop('Unknown', errors='ignore')
print("Top 10 Réalisateurs (Nombre de Titres):")
print(director_counts)

# Visualisation (Seaborn)
plt.figure(figsize=(12, 6))
sns.barplot(x=director_counts.index, y=director_counts.values, hue=director_counts.index, palette='RdPu', legend=False)
plt.title('Top 10 des Réalisateurs les Plus Fréquents', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.show()

# --- TOP 10 ACTEURS/ACTRICES ---
cast_counts = all_cast.value_counts().head(10).drop('Unknown', errors='ignore')
print("\nTop 10 Acteurs/Actrices (Nombre de Titres):")
print(cast_counts)

# Visualisation (Seaborn)
plt.figure(figsize=(12, 6))
sns.barplot(x=cast_counts.index, y=cast_counts.values, hue=cast_counts.index, palette='YlGnBu', legend=False)
plt.title('Top 10 des Acteurs/Actrices les Plus Fréquents', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.show()
```
